```
data = spark.read.csv('FileStore/tables/HR_comma_sep_csv.txt',header=True, inferSchema=True)
```

```
display(data)
```

```
data = data.withColumnRenamed('sales','dept')
```

```
display(data)
```

```
cat_col_names = ['dept','salary']
```

```
from pyspark.ml.feature import StringIndexer
```

```
si = StringIndexer(inputCol = 'dept',outputCol = 'dept_en')
```

```
si_model = si.fit(data)
```

```
display(si_model.transform(data))
```

```
for col in cat_col_names:
  si = StringIndexer(inputCol = col, outputCol = col+'_en')
  si_model = si.fit(data)
  data = si_model.transform(data)
  
```

```
display(data)
```

```
num_col_names = ['number_project', 'average_montly_hours', 'time_spend_company']
```

```
from pyspark.ml.feature import VectorAssembler
```

```
ve = VectorAssembler(inputCols = num_col_names, outputCol = 'vec_ss')
```

```
data = ve.transform(data)
```

```
from pyspark.ml.feature import StandardScaler
```

```
ss = StandardScaler(inputCol = 'vec_ss', outputCol = 'vec_ss_x')
ss_model = ss.fit(data)
data = ss_model.transform(data)
```

```
display(data)
```

```
ss = StandardScaler(inputCol = 'number_project', outputCol='p')
```

```
ve = VectorAssembler(inputCols = ['satisfaction_level','last_evaluation', 'dept_en','salary_en'],outputCol='vec_nn')
```

```
data = ve.transform(data)
```

```
display(data)
```

```
ve = VectorAssembler(inputCols=['vec_nn','vec_ss_x'],outputCol='feature')
```

```
data = ve.transform(data)
```

```
display(data)
```

```
from pyspark.ml.classification import LogisticRegression
```

```
lr = LogisticRegression(featuresCol='feature',labelCol='left')
```

```
lr_model = lr.fit(data)
```

```
display(lr_model.transform(data))
```

```
train, test = data.randomSplit([.7,.3])
```

```
lr_model = lr.fit(train)
```

```
test_o  = lr_model.transform(test)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
bce = BinaryClassificationEvaluator(labelCol='left')
```

```
bce.evaluate(test_o)
```
